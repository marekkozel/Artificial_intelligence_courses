```python
import matplotlib.pyplot as plt # Plotting only
import numpy as np
```

```python
# Second one is a triangular pulse.

def sig_square(x):
  return 0 if x < 3 or x > 5 else 2


def sig_triag(x):
  return 0 if x < 0 or x > 2 else x

# First signal (square pulse)
sig1 = [sig_square(x/10) for x in range(100)]


# Second signal (triangle pulse)
sig2 = [sig_triag(x/10) for x in range(20)]


plt.figure(figsize = (5,5))
plt.subplot(1,1,1)
plt.plot(sig1)

plt.figure(figsize = (10,7))
plt.subplot(1,2,2)
plt.plot(sig2)
```

```python
sig3 = [sig_square(x/2) for x in range(20)]
plt.plot(np.multiply(np.array(sig2), np.array(sig3)))
print(np.multiply(np.array(sig2), np.array(sig3)))
```

```python
''' Python only implementation '''


# Pre-allocate correlation array
conv = (len(sig1) - len(sig2)) * [0]
print(conv)
print(len(conv))
print(len(sig2))
# Go through lag components one-by-one
for l in range(len(conv)):
  for i in range(len(sig2)):
    conv[l] += sig1[l-i+len(sig2)] * sig2[i]

  conv[l] /= len(sig2) # Normalize

plt.plot(conv)
print(conv)
```
